```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
p = nx.DiGraph()
nodes = ['A', 'B', 'C', 'D', 'E']
p.add_nodes_from(nodes)
edges = [('A', 'B'), ('B', 'C'), ('C', 'A'), ('A', 'D'), ('D', 'E'), ('E', 'D'), ('B', 'E'), ('E', 'B')]
p.add_edges_from(edges)
```

```python
degrees = {}
for node in nodes:
    degrees[node] = {
        'in_degree': p.in_degree(node),
        'out_degree': p.out_degree(node)
    }
```

```python
pos = nx.spring_layout(p)
nx.draw_networkx_nodes(p, pos, node_color='lightblue')
nx.draw_networkx_edges(p, pos, edgelist=edges, arrowstyle='->')
nx.draw_networkx_labels(p, pos,font_color='darkblue')
plt.title("Directed Graph")
plt.axis('off')
plt.show()
```
