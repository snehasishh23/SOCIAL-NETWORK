```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
p=nx.Graph()
p
```

```python
p.add_weighted_edges_from([
    ("A", "B", 5),
    ("A", "C", 3),
    ("B", "C", 4),
    ("B", "D", 6),
    ("C", "E", 2),
    ("D", "E", 7),
    ("A", "E", 1)
])
```

```python
degree_cen = nx.degree_centrality(p)
```

```python
weighted_degree_cen = {}
for node in p.nodes():
    edges_with_data = p.edges(node, data=True)
    total_weight = sum(data['weight'] for _, _, data in edges_with_data)
    weighted_degree_cen[node] = total_weight
```

```python
pos = nx.spring_layout(p)
plt.figure(figsize=(8, 6))
nx.draw_networkx_nodes(p, pos, node_size=2000, node_color="lightblue")
edges, weights = zip(*nx.get_edge_attributes(p, 'weight').items())
nx.draw_networkx_edges(p, pos, edgelist=edges, width=[weight for weight in weights], edge_color="grey")
nx.draw_networkx_labels(p, pos, font_size=12, font_color="black")
nx.draw_networkx_edge_labels(p, pos, edge_labels=nx.get_edge_attributes(p, 'weight'))
plt.title("Weighted Graph")
plt.show()
```

```python
for node, centrality in weighted_degree_cen.items():
    print(f"Node: {node}, Weighted Degree Centrality: {centrality}")
```
